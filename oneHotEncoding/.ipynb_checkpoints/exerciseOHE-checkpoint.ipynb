```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
final = merged.drop(['Car Model', 'Audi A5'], axis='columns')
final
```

```python
%matplotlib inline
plt.scatter(df['Mileage'], df['Sell Price($)'])
```

```python
X = final.drop('Sell Price($)', axis='columns')
X
```

```python
model = linear_model.LinearRegression()
model.fit(X,final['Sell Price($)'])
```

```python
model.predict([[45000, 4, 0, 1]])
```

```python
model.predict([[86000, 7, 1, 0]])
```

```python
y=final['Sell Price($)']
model.score(X,y)
```

```python
dfle = df
dfle
```

```python
le = LabelEncoder()
dfle['Car Model'] = le.fit_transform(dfle['Car Model'])
dfle
```

```python
X = dfle[['Car Model', 'Mileage', 'Age(yrs)']]
X
```

```python
y = dfle['Sell Price($)']
y
```

```python
ohe = ColumnTransformer([('ohe', OneHotEncoder(), [0])], remainder='passthrough')
```

```python
X = ohe.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model_new = linear_model.LinearRegression()
model_new.fit(X,y)
```

```python
model_new.predict([[0, 1, 45000, 4]])
```

```python
model_new.predict([[1, 0, 86000, 7]])
```

```python
model_new.score(X,y)
```
